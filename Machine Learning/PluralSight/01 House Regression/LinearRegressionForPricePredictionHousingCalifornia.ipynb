```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
housing_data = pd.read_csv('datasets/housing.csv')
#The difference compared to head method is that sample is random
housing_data.sample(5)
```

```python
#drop data with missing field
housing_data = housing_data.dropna()
```

```python
#number of (entry,variable) after dropping empty entries
housing_data.shape
```

```python
#this loc method is to count the number of entry that have certain value
#as you can see, there are 958 entries with median_house_value = 500001,
housing_data.loc[housing_data['median_house_value'] ==500001].count()
```

```python
#This upper value of house value 500001 will scatter our data model, thus, we need to remove using drop method
housing_data = housing_data.drop(housing_data.loc[housing_data['median_house_value'] == 500001].index)
```

```python
housing_data.shape
```

```python
housing_data.head()
```

```python
#ocean_proximity have discrete value, we need to convert
housing_data['ocean_proximity'].unique()
```

```python
#using one hot encoding to numeric from using get_dumm from panda
housing_data = pd.get_dummies(housing_data, columns=['ocean_proximity'])
```

```python
#now it have 14 variable. 10-1+5 . 5 is 5 kind of value in ocean proximity
housing_data.shape
```

```python
#discrete variable of ocean_proximity
housing_data.sample(5)
```

```python
#now start to set feature and target in linear regression
#X is X axis, but dropping median_house_value because this is our target
#Y axis is median house value only
X = housing_data.drop('median_house_value', axis=1)
Y = housing_data['median_house_value']
```

```python
X.columns #without median_house_income
```

```python
#split dataset into test and train datasets. Train is to train the model, test is to test the model, its also shuffling
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2) #test is 20%, train is 80%
```

```python
x_train.shape, x_test.shape
```

```python
y_train.shape, y_test.shape
```

```python
#start the linear regression
from sklearn.linear_model import LinearRegression
#normalization scale all numerics in 0-1, greatly improve data processing speed. (preprocessing)
linear_model = LinearRegression(normalize=True).fit(x_train,y_train)
```

```python
#For regression, this is R^2
print("Training_score : " , linear_model.score(x_train, y_train))
```

```python
predictors = x_train.columns
predictors #is list of coloumn of features
```

```python
#calculate coefficient of all features using Panda Series
#as you can see, more + coef means that it contribute much to median_house_price (Y axis)
coef = pd.Series(linear_model.coef_,predictors).sort_values()
print(coef)
```

```python
#to predict the value (median_house_value) from x_test
y_pred = linear_model.predict(x_test)
```

```python
#dataframe that compare predicted value with actual value
df_pred_actual = pd.DataFrame({'predicted' : y_pred, 'actual' : y_test})
df_pred_actual.sample(10)
```

```python
#evaluate a model , calculate R2 score for test data
from sklearn.metrics import r2_score

print("Testing Score: ", r2_score(y_test, y_pred))#first input is actual value, 2nd input is predicted value
```

```python
#use scatter plot to visualize predicted vs actual in test series

fig, ax = plt.subplots(figsize=(12,8))
plt.scatter(y_test,y_pred)
plt.show()
```

```python
#sample 100
df_pred_actual_sample = df_pred_actual.sample(100)
df_pred_actual_sample = df_pred_actual_sample.reset_index()
```
