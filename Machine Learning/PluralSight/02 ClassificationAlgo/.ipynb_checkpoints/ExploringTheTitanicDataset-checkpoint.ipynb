```python
import sklearn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#SibSp is sibling and Spous, Parch is parent and child, Pclass = Passanger class (highest 1), embarked is sections
titanic_df = pd.read_csv('datasets/titanic_train.csv')
titanic_df.head(10)
```

```python
titanic_df.shape
```

```python
#drop information that is useless
titanic_df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], 'columns', inplace=True)
titanic_df.head()
```

```python
#Check missing fields in dataset
titanic_df[titanic_df.isnull().any(axis=1)].count()
```

```python
#Drop record if having even only 1 field missing
titanic_df = titanic_df.dropna()
titanic_df.shape
```

```python
#see statistics of each variable
titanic_df.describe()
```

```python
#plot to see relationship between variable
fig, ax = plt.subplots(figsize=(12,8))
plt.scatter(titanic_df['Age'], titanic_df['Survived'])
plt.xlabel('Age')
plt.ylabel('Survived')
```

```python
#plot to see relationship between variable
fig, ax = plt.subplots(figsize=(12,8))
plt.scatter(titanic_df['Fare'], titanic_df['Survived'])
plt.xlabel('Fare')
plt.ylabel('Survived')
```

```python
#Sex is not value, we can see female have higher chance to survive
pd.crosstab(titanic_df['Sex'], titanic_df['Survived'])
```

```python
#we can see that Class 1 have higher chance of survive, compared to lower class
pd.crosstab(titanic_df['Pclass'], titanic_df['Survived'])
```

```python
#preprocessing sex. convert into numeric value
from sklearn import preprocessing
label_encoding = preprocessing.LabelEncoder()
titanic_df['Sex'] = label_encoding.fit_transform(titanic_df['Sex'].astype(str))
titanic_df.head()
```

```python
#female is 0, male is 1
label_encoding.classes_
```

```python
#preprocessing Embarked using One Hot Encoding
titanic_df = pd.get_dummies(titanic_df, columns=['Embarked'])
titanic_df.head()
```

```python
#view correlation
#Sex, female = 0 , so it have highest negative correlation
titanic_data_corr = titanic_df.corr()
titanic_data_corr
```

```python
#heatmap correlation
fig, ax = plt.subplots(figsize=(12, 10))
sns.heatmap(titanic_data_corr, annot=True)
```

```python
#frac = 1 mean entire datasets
#This code only shuffling the training data
titanic_df = titanic_df.sample(frac=1).reset_index(drop=True)
titanic_df.head()
```

```python
titanic_df.to_csv('datasets/titanic_procesed.csv', index=False) #no indexing
```
