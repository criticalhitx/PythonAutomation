```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
automobile_df = pd.read_csv('datasets/auto-mpg-processed.csv')
automobile_df.sample(5)
```

```python
from sklearn.model_selection import train_test_split

X = automobile_df[['horsepower']]
Y = automobile_df['mpg']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
x_train.sample(5)
```

```python
#normalizing data is standard preprocessing for LinReg
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression(normalize=True).fit(x_train,y_train)
```

```python
print('Training score: ', linear_model.score(x_train, y_train))
```

```python
y_pred = linear_model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
print('Testing score: ', r2_score(y_test, y_pred))
```

```python
fig, ax = plt.subplots(figsize=(12,8))

plt.scatter(x_test, y_test)
plt.plot(x_test, y_pred, color = 'r')

plt.xlabel('Horsepower')
plt.ylabel('Mpg')
plt.show()
```

```python
#another feature = age
X = automobile_df[['age']]
Y = automobile_df['mpg']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)

linear_model = LinearRegression(normalize=True).fit(x_train,y_train)

print('Training score: ', linear_model.score(x_train, y_train))

y_pred = linear_model.predict(x_test)

print('Testing score: ', r2_score(y_test, y_pred))
```

```python
# you can see the data is greatly splited
fig, ax = plt.subplots(figsize=(12,8))

plt.scatter(x_test, y_test)
plt.plot(x_test, y_pred, color = 'r')

plt.xlabel('Age')
plt.ylabel('Mpg')
plt.show()
```

```python
X = automobile_df[['displacement', 'horsepower', 'weight','acceleration', 'cylinders']]
Y = automobile_df['mpg']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
linear_model = LinearRegression(normalize=True).fit(x_train, y_train)
```

```python
print('Training score: ', linear_model.score(x_train, y_train))
```

```python
predictors = x_train.columns
coef = pd.Series(linear_model.coef_, predictors).sort_values()

print(coef)
#all negative which means when these value go up, mpg goes down
```

```python
y_pred = linear_model.predict(x_test)
```

```python
print('Testing score: ', r2_score(y_test, y_pred))
```

```python
plt.figure(figsize = (20,10))

plt.plot(y_pred, label='Predicted')
plt.plot(y_test.values, label='actual')
plt.ylabel('Mpg')
plt.legend()
plt.show()
```
