```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
exam_df = pd.read_csv('datasets/exams.csv')
exam_df.head()
```

```python
exam_df.describe()
```

```python
#this data need to be preprocessed a lot
#scale have 0 mean and unit variance
from sklearn import preprocessing
exam_df[['math score']] = preprocessing.scale(exam_df[['math score']].astype('float64'))
exam_df[['reading score']] = preprocessing.scale(exam_df[['reading score']].astype('float64'))
exam_df[['writing score']] = preprocessing.scale(exam_df[['writing score']].astype('float64'))
```

```python
exam_df.head()
```

```python
#mean near 0, std near 1
exam_df.describe()
```

```python
exam_df['parental level of education'].unique()
```

```python
#because parental level of education have ordering, use this labeling
parent_level_of_education = [
    "some high school",
    "high school",
    "some college",
    "associate's degree",
    "bachelor's degree",
    "master's degree"]
```

```python
label_encoding = preprocessing.LabelEncoder()
label_encoding = label_encoding.fit(parent_level_of_education)
```

```python
exam_df['parental level of education'] =label_encoding.transform(exam_df['parental level of education']).astype(str)
```

```python
#parent education has been rearranged
exam_df.head()
```

```python
#to see label
label_encoding.classes_
```

```python
#ethnic/race doesnt have intrinsic order, so we can use one hot encoding
exam_df = pd.get_dummies(exam_df, columns=['race/ethnicity'])

exam_df.head()
```

```python
#use another for one hot encoding
exam_df = pd.get_dummies(exam_df, columns=['gender','lunch','test preparation course'])
```

```python
exam_df.head()
```

```python
from sklearn.model_selection import train_test_split

X = exam_df.drop('math score', axis=1)
Y = exam_df['math score']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
x_train.shape, y_train.shape
```

```python
x_test.shape, y_test.shape
```

```python
#we've use one hot encoding for features and have explicitly set fit_intercept=True - This mimght cause us to encounter perfect collinearity in our model
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression(fit_intercept=True).fit(x_train, y_train)
```

```python
print("Training_score : ", linear_model.score(x_train, y_train))
```

```python
y_pred = linear_model.predict(x_test)

from sklearn.metrics import r2_score
print("Testing_score : ", r2_score(y_test, y_pred))
```

```python
#now we set fit_intercept to false
#when we use all column in hot encoding, fit_intercept should be False
linear_model = LinearRegression(fit_intercept=False).fit(x_train, y_train)
```

```python
print("Training_score : ", linear_model.score(x_train, y_train))
```

```python
y_pred = linear_model.predict(x_test)

from sklearn.metrics import r2_score
print("Testing_score : ", r2_score(y_test, y_pred))
```

```python
#Why the result is the same? dummy variable trap usually happen when
#fit intercept = True. Scikit-learn doesnt care if u use hot encoding,
#scikit will force you to use fit_intercept = false, so you wont fall into
#dummy trap
```

```python
df_pred_actual = pd.DataFrame({'predicted': y_pred, 'actual': y_test})
df_pred_actual.head(10)
```

```python
#plot line chart actual vs predicted

plt.figure(figsize= (12,8))

plt.plot(y_pred, label='Predicted')
plt.plot(y_test.values, label='Actual')

plt.ylabel('math_score')
plt.legend()
plt.show()
```

```python
# we want to try without another scores
X = exam_df.drop(['math score', 'writing score',  'reading score'], axis=1)
Y = exam_df['math score']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
linear_model = LinearRegression(fit_intercept=False).fit(x_train, y_train)
print("Training_score : ", linear_model.score(x_train, y_train))

y_pred = linear_model.predict(x_test)
print("Testing_score : ", r2_score(y_test, y_pred))
```

```python
#why ? because other scores that have highest predictive power
```
